**Creating Numpy Arrays:**
There are two common ways to create a numpy array.


*   Creating from existing python lists.
*   Creating from scratch




```
test_arr = np.array([[1,2],[3,4]])
test_arr.shape
```

**Creating from scratch:**

1.   np.zeros() -  create a numpy array of particular dimension with just zeros in it. 
2.   np.arange() - create a numpy array of particular dimension with range of values in it




```
zero_arr = np.zeros((2, 3))
zero_arr
```

```
arange_arr = np.arange(4, 10, dtype=np.integer)
arange_arr
```

**Accessing numpy arrays:**
You can access numpy arrays using indices. Index starts with 0.

```
# 1. Use one index access for each index. For example, test_arr[0][1] will give you first row and second column. Two different syntax to do the same thing.
print(test_arr[0][1])
print(test_arr[0,1])

# 2. giving parent dimension will fetch you all children for that parent index. For example, test_arr[0] will give you the entire first row.
print(test_arr[0])

# 3. if you want the first colum, use this ":" operator.":" means all values.
print(test_arr[:][1])

# 4. This works but its negligible
print(test_arr[0][:])
```

**Slicing numpy arrays:**
When you work with huge datasets, its easy to work with a subset of data. Slicing comes in handy in those situations!

```
num_array = np.arange(10)

#slice items starting from index 
arr1 = num_array[4:]
print(arr1)

#slice items ending in an index from start
arr2 = num_array[:4]
print(arr2)

#slice items with start and end index
arr2 = num_array[2:4]
print(arr2)

#Note: Start index inclusive and end index exclusive
```

You can slice numpy array of any dimensions.

```
num_array_nd = np.array([[1,2,3,3],[4,5,6,6],[7,8,9,9]])

#To get rows
first_row = num_array_nd[0,...]
print(first_row)

first_row_1 = num_array_nd[0][:]
print(first_row_1)

#To get columns
second_column = num_array_nd[...,1]
print(second_column)

#Slice by range
print(num_array_nd[1:3])
```

**Note**
There are a few things to watch out for when you are slicing. Slicing is just a reference to a portion of parent array. If you modify the slice, the parent gets modified too! In such cases use np.array(slice).

```
parent_arr = np.arange(10)
print(parent_arr)
sliced = parent_arr[2:5]
print(sliced)
sliced[0] = 100
print(sliced)
print(parent_arr)

#using np.array
parent_arr = np.arange(10)
print(parent_arr)
sliced = np.array(parent_arr[2:5])
print(sliced)
sliced[0] = 100
print(sliced)
print(parent_arr)
```

There are multiple ways to index and slice a numpy array. You can refer them here https://docs.scipy.org/doc/numpy-1.15.1/reference/arrays.indexing.html 

**Handy Utilities to Know:**

*   np.shape - gives the dimensions of a numpy array
*   np.reshape() - reshape dimensions of a numpy array. Comes in handy when you process image datasets.




```
num_array_nd = np.array([[1,2,3,10],[4,5,6,11],[7,8,9,12]])

#shape of an array
print(num_array_nd.shape)

#reshaping a numpy array
reshaped_arr = num_array_nd.reshape(4,3)
print(reshaped_arr)
```

## Matplotlib: A Python Library for Data Visualization


**MATPLOTLIB**
It is a Python 2D-plotting library.
We can generate graphs, histograms and bar plots using the library functions.

### Simple 2D Plots with Matplotlib

Matplotlib's plot function allows us to create simple 2D line plots. Let's try to create a line plot visualizing the trip distance versus the trip duration.


```
plt.plot(df['google_distance'], df['google_duration'])
plt.title('Distance vs. Duration')
plt.xlabel('Distance')
plt.ylabel('Trip Duration')
plt.show()
```

The plot above tells us that in general there is a positive correlation between distance and the trip duration but it looks a little messy. Perhaps a scatterplot would look better.

### Scatter Plots with Matplotlib
Let's repeat the same example above but this time we will create a scatter plot.

```
plt.scatter(df['google_distance'], df['google_duration'])
plt.title('Distance vs. Duration')
plt.xlabel('Distance')
plt.ylabel('Trip Duration')
plt.show()
```

```
latitude = df['pickup_latitude']
latitude_arr = latitude.values

plt.plot(latitude_arr)

plt.xlabel('data records')
plt.ylabel('pickup_latitude')

plt.title("Simple Plot - Latitude")


plt.show()

```

### Histograms Using Matplotlib


```
plt.hist(df['passenger_count'], bins=5, alpha=0.5, facecolor='red')
plt.show()
```
